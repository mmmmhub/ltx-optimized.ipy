## Architecture Overview

| Component | Detail |
|-----------|--------|
| **Default GPU** | H200 141GB SXM |
| **CPU** | 8 Cores |

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 1 — CONFIGURATION & CONSTANTS                 ║
# ╚══════════════════════════════════════════════════════════════╝

import os

# ── GPU target (configurable — no hardcoded GPU) ──────────────
# Override via environment variable: GPU_TYPE=H100 modal deploy ...
# Valid values: "H200", "H100", "A100", "any"
GPU_TYPE = os.environ.get("GPU_TYPE", "H200").upper()

# ── Paths inside the Modal container ──────────────────────────
MODELS_DIR   = "/models"
OUTPUTS_DIR  = "/outputs"
WAN2GP_DIR   = "/wan2gp"
GEMMA_FOLDER = "gemma-3-12b-it-qat-q4_0-unquantized"

# ── Model identifiers ─────────────────────────────────────────
HF_TRANSFORMER_REPO = "Abiray/LTX-2.3-22B-DISTILLED-1.1-GGUF"
HF_COMPANION_REPO   = "DeepBeepMeep/LTX-2"
TRANSFORMER_FILE    = "LTX-2.3-22B-distilled-1.1-Q4_K_M.gguf"
TRANSFORMER_DEST    = "ltx-2.3-22b-distilled-1.1-Q4_K_M.gguf"  # lowercase for Wan2GP

COMPANION_LARGE = [
    "ltx-2.3-22b-distilled-lora-384.safetensors",
    "ltx-2.3-22b_embeddings_connector.safetensors",
    "ltx-2.3-22b_text_embedding_projection.safetensors",
    "ltx-2.3-22b_vae.safetensors",
]

COMPANION_SMALL = [
    "ltx-2.3-22b_audio_vae.safetensors",
    "ltx-2.3-22b_vocoder.safetensors",
    "ltx-2.3-spatial-upscaler-x2-1.1.safetensors",
    "ltx-2.3-temporal-upscaler-x2-1.0.safetensors",
]

GEMMA_FILES = [
    "gemma-3-12b-it-qat-q4_0-unquantized_quanto_bf16_int8.safetensors",
    "added_tokens.json", "chat_template.json", "config_light.json",
    "generation_config.json", "preprocessor_config.json", "processor_config.json",
    "special_tokens_map.json", "tokenizer.json", "tokenizer.model", "tokenizer_config.json",
]

# ── Default generation parameters ─────────────────────────────
# These are hardware-agnostic; VRAM-dependent flags are set at runtime
DEFAULT_WIDTH      = 1920
DEFAULT_HEIGHT     = 1088   # 1080p snapped to 32-multiple
DEFAULT_FPS        = 24.0
DEFAULT_FRAMES     = 481    # ~20 seconds @ 24fps (8k+1 valid LTX frame count)
DEFAULT_GUIDE      = 3.0
DEFAULT_STEPS      = 8

# ── VRAM thresholds for automatic mode selection ───────────────
# >120 GB  → HIGH_PERF  (H200 141GB)   — maximum VRAM budgets, no offloading
# >60 GB   → BALANCED   (H100/A100 80GB) — generous budgets, minimal offloading
# <=60 GB  → EFFICIENT  (RTX PRO 6000 etc.) — conservative budgets
VRAM_HIGH_PERF_THRESHOLD_GB  = 120.0   # H200 141GB detection threshold
VRAM_BALANCED_THRESHOLD_GB   =  60.0   # H100/A100 80GB detection threshold

# ── CUDA allocator / attention flags ──────────────────────────
CUDA_ALLOC_CONF = "expandable_segments:True,garbage_collection_threshold:0.8"
ATTENTION_MODE  = "sdpa"   # Flash Attention 2 via torch SDPA (sm_80+ / sm_90+)

# ── Modal volume / app names ───────────────────────────────────
MODAL_APP_NAME    = "ltx-23-22b-h200"
MODAL_VOLUME_NAME = "ltx-23-22b-models"
MODAL_OUTPUT_VOL  = "ltx-23-22b-outputs"

# ── Output cleanup: keep last N videos ─────────────────────────
MAX_OUTPUTS_KEEP  = 50

print("✅ Configuration loaded.")
print(f"   Target GPU         : {GPU_TYPE}")
print(f"   Default resolution : {DEFAULT_WIDTH}×{DEFAULT_HEIGHT}")
print(f"   Default frames     : {DEFAULT_FRAMES} ({DEFAULT_FRAMES/DEFAULT_FPS:.1f}s @ {DEFAULT_FPS}fps)")
print(f"   Models volume      : {MODAL_VOLUME_NAME}")
print(f"   Outputs volume     : {MODAL_OUTPUT_VOL}")
print(f"   VRAM high-perf threshold : >{VRAM_HIGH_PERF_THRESHOLD_GB} GB → HIGH_PERF mode")
print(f"   VRAM balanced threshold  : >{VRAM_BALANCED_THRESHOLD_GB} GB → BALANCED mode")
````

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 2 — MODAL APP, IMAGE, VOLUMES, GPU CONFIG     ║
# ╚══════════════════════════════════════════════════════════════╝

import modal

# ── Resolve Modal GPU object from GPU_TYPE string ──────────────
# Supports: H200, H100, A100, any
def _resolve_modal_gpu(gpu_type: str):
    """Return the appropriate modal.gpu object for the requested GPU_TYPE.
    Falls back to H200 if the string is unrecognised.
    """
    t = gpu_type.upper().strip()
    if t == "H200":
        return modal.gpu.H200(count=1)
    elif t == "H100":
        return modal.gpu.H100(count=1)
    elif t in ("A100", "A100_80GB"):
        return modal.gpu.A100(count=1, size="80GB")
    elif t == "ANY":
        return "any"
    else:
        # Unrecognised type — default to H200 and warn
        print(f"[CONFIG] ⚠️  Unknown GPU_TYPE '{gpu_type}', defaulting to H200.")
        return modal.gpu.H200(count=1)

MODAL_GPU = _resolve_modal_gpu(GPU_TYPE)

# ── Modal Volumes ──────────────────────────────────────────────
model_volume  = modal.Volume.from_name(MODAL_VOLUME_NAME,  create_if_missing=True)
output_volume = modal.Volume.from_name(MODAL_OUTPUT_VOL,   create_if_missing=True)

# ── Container image ────────────────────────────────────────────
# PyTorch 2.5.1+cu124  — compatible with sm_80 (A100), sm_90 (H100/H200)
# and sm_89 (RTX PRO 6000).  CUDA 12.4 supports all four targets.
ltx_image = (
    modal.Image.from_registry(
        "nvidia/cuda:12.4.1-cudnn-devel-ubuntu22.04",
        add_python="3.11",
    )
    .apt_install(
        "git", "ffmpeg", "libsndfile1", "wget", "curl",
        "libgl1-mesa-glx", "libglib2.0-0",
    )
    .pip_install(
        "torch==2.5.1",
        "torchvision==0.20.1",
        "torchaudio==2.5.1",
        extra_index_url="https://download.pytorch.org/whl/cu124",
    )
    .run_commands(
        "git clone https://github.com/DeepBeepMeep/Wan2GP.git /wan2gp",
        "pip install --timeout 180 --retries 5 -r /wan2gp/requirements.txt",
        "pip install --timeout 180 --retries 5 mmgp gguf soundfile huggingface_hub fastapi uvicorn pydantic psutil",
    )
)

# ── Modal App ──────────────────────────────────────────────────
app = modal.App(
    name=MODAL_APP_NAME,
    image=ltx_image,
)

print("✅ Modal app, image and volumes configured.")
print(f"   App name    : {MODAL_APP_NAME}")
print(f"   GPU target  : {GPU_TYPE}  →  {MODAL_GPU}")
print(f"   Base image  : nvidia/cuda:12.4.1-cudnn-devel-ubuntu22.04 + PyTorch 2.5.1+cu124")
print(f"   Model volume: {MODAL_VOLUME_NAME}  →  {MODELS_DIR}")
print(f"   Output vol  : {MODAL_OUTPUT_VOL}   →  {OUTPUTS_DIR}")
````

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 3 — MODEL DOWNLOAD & VOLUME PERSIST           ║
# ╚══════════════════════════════════════════════════════════════╝

@app.function(
    volumes={
        MODELS_DIR:  model_volume,
        OUTPUTS_DIR: output_volume,
    },
    timeout=7200,        # 2 hours — large downloads
    cpu=4,
    memory=16384,
)
def download_models():
    """Download all required model files into the persistent Modal Volume.
    Skips files that already exist.  Call once with `modal run` before deploying.
    """
    import os
    from huggingface_hub import hf_hub_download

    models_dir   = MODELS_DIR
    gemma_dir    = os.path.join(models_dir, GEMMA_FOLDER)
    os.makedirs(models_dir, exist_ok=True)
    os.makedirs(os.path.join(models_dir, "ckpts"), exist_ok=True)
    os.makedirs(OUTPUTS_DIR, exist_ok=True)

    def _dl(repo, filename, dest_path):
        if os.path.isfile(dest_path):
            size_mb = os.path.getsize(dest_path) / 1024**2
            print(f"  ✓ EXISTS  {os.path.basename(dest_path)} ({size_mb:.0f} MB)")
            return
        print(f"  ↓ DOWNLOADING  {filename}  from  {repo} ...")
        tmp = hf_hub_download(repo_id=repo, filename=filename, local_dir=models_dir)
        # hf_hub_download may place file under subdir; move if needed
        if os.path.abspath(tmp) != os.path.abspath(dest_path):
            os.makedirs(os.path.dirname(dest_path), exist_ok=True)
            os.rename(tmp, dest_path)
        size_mb = os.path.getsize(dest_path) / 1024**2
        print(f"  ✓ DONE     {os.path.basename(dest_path)} ({size_mb:.0f} MB)")

    print("\n=== Downloading LTX-2.3 22B Distilled 1.1 Q4_K_M Transformer ===")
    _dl(
        HF_TRANSFORMER_REPO,
        TRANSFORMER_FILE,
        os.path.join(models_dir, TRANSFORMER_DEST),
    )

    print("\n=== Downloading Companion Large Files ===")
    for f in COMPANION_LARGE:
        _dl(HF_COMPANION_REPO, f, os.path.join(models_dir, f))

    print("\n=== Downloading Small Model Files ===")
    for f in COMPANION_SMALL:
        _dl(HF_COMPANION_REPO, f, os.path.join(models_dir, f))

    print("\n=== Downloading Gemma-3 12B Text Encoder ===")
    os.makedirs(gemma_dir, exist_ok=True)
    for gf in GEMMA_FILES:
        _dl(
            HF_COMPANION_REPO,
            f"{GEMMA_FOLDER}/{gf}",
            os.path.join(gemma_dir, gf),
        )

    # ── Verify all required files present ──────────────────────
    print("\n=== Verifying All Required Files ===")
    required = (
        [os.path.join(models_dir, TRANSFORMER_DEST)]
        + [os.path.join(models_dir, f) for f in COMPANION_LARGE]
        + [os.path.join(models_dir, f) for f in COMPANION_SMALL]
        + [os.path.join(gemma_dir,  gf) for gf in GEMMA_FILES]
    )
    missing = [p for p in required if not os.path.isfile(p)]
    if missing:
        raise RuntimeError(f"Missing files after download:\n" + "\n".join(missing))
    print(f"  ✅ All {len(required)} required files verified.")

    # Commit so Volume reflects changes
    model_volume.commit()
    output_volume.commit()
    print("\n✅ Model download complete and Volume committed.")


print("✅ download_models() Modal function defined.")
print("   Run with:  modal run ltx_modal_h200.ipynb::download_models")
````

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 4 — MODEL LOADING (SINGLETON @modal.enter)    ║
# ╚══════════════════════════════════════════════════════════════╝
#
# The LTXVideoService class holds ALL model state.
# @modal.enter() fires ONCE per container lifetime → models stay
# resident in VRAM/RAM for the container lifetime.
#
# VRAM auto-detection at startup:
#   >120 GB  → HIGH_PERF  (H200 141GB)  — maximum budgets, no offloading
#   >60 GB   → BALANCED   (H100/A100 80GB) — generous budgets
#   <=60 GB  → EFFICIENT  (RTX PRO 6000 etc.) — conservative budgets

import modal
import json

@app.cls(
    gpu=MODAL_GPU,
    cpu=8,
    memory=65536,          # 64 GB RAM
    timeout=3600,          # 1-hour request timeout
    volumes={
        MODELS_DIR:  model_volume,
        OUTPUTS_DIR: output_volume,
    },
    concurrency_limit=1,   # one generation at a time per container
    allow_concurrent_inputs=False,
)
class LTXVideoService:
    """Singleton Modal class for LTX-2.3 22B.
    Automatically adapts memory budgets to detected VRAM:
      HIGH_PERF  (>120 GB)  → H200 141GB optimised
      BALANCED   (>60 GB)   → H100 / A100 80GB optimised
      EFFICIENT  (<=60 GB)  → RTX PRO 6000 / smaller cards
    """

    @modal.enter()
    def load_models(self):
        """Called exactly once when the container starts.
        Loads Wan2GP + mmgp pipeline and keeps everything in memory.
        """
        import gc
        import os
        import sys
        import glob
        import time
        import psutil
        import torch

        t0 = time.time()

        # ── Environment flags ──────────────────────────────────
        os.environ["PYTORCH_CUDA_ALLOC_CONF"]  = CUDA_ALLOC_CONF
        os.environ["TOKENIZERS_PARALLELISM"]    = "false"
        os.environ["CUDA_LAUNCH_BLOCKING"]      = "0"

        # ── Bootstrap Wan2GP from volume-mounted clone ─────────
        sys.path.insert(0, WAN2GP_DIR)
        os.chdir(WAN2GP_DIR)

        # ── Hardware reporting ─────────────────────────────────
        assert torch.cuda.is_available(), "No CUDA device detected!"
        gpu_name   = torch.cuda.get_device_name(0)
        gpu_sm     = torch.cuda.get_device_capability(0)
        vram_total = torch.cuda.get_device_properties(0).total_memory / 1024**3
        ram_gb     = psutil.virtual_memory().total / 1024**3
        cuda_ver   = torch.version.cuda or "unknown"
        torch_ver  = torch.__version__

        print("\n" + "=" * 70)
        print("[HARDWARE REPORT]")
        print(f"  GPU Name     : {gpu_name}")
        print(f"  VRAM         : {vram_total:.1f} GB  (sm_{gpu_sm[0]}{gpu_sm[1]})")
        print(f"  RAM          : {ram_gb:.1f} GB")
        print(f"  CUDA Version : {cuda_ver}")
        print(f"  PyTorch      : {torch_ver}")
        print("=" * 70 + "\n")

        # sm_80 = A100, sm_89 = RTX PRO 6000 / L40, sm_90 = H100/H200
        assert gpu_sm[0] >= 8, (
            f"Requires Ampere or newer (sm_80+). Detected sm_{gpu_sm[0]}{gpu_sm[1]}"
        )

        # ── Automatic VRAM-based mode selection ────────────────
        if vram_total > VRAM_HIGH_PERF_THRESHOLD_GB:
            vram_mode = "HIGH_PERF"
        elif vram_total > VRAM_BALANCED_THRESHOLD_GB:
            vram_mode = "BALANCED"
        else:
            vram_mode = "EFFICIENT"

        print(f"[INIT] VRAM mode : {vram_mode}  ({vram_total:.1f} GB detected)")
        self._vram_mode  = vram_mode
        self._vram_total = vram_total

        # ── SDPA / attention flags ─────────────────────────────
        torch.backends.cuda.enable_flash_sdp(True)
        torch.backends.cuda.enable_mem_efficient_sdp(True)
        torch.backends.cuda.enable_math_sdp(True)

        # ── Import Wan2GP pipeline components ─────────────────
        from mmgp import offload
        from shared.utils import files_locator as fl
        fl.set_checkpoints_paths([MODELS_DIR, os.path.join(MODELS_DIR, "ckpts"), "."])

        from models.ltx2.ltx2_handler import family_handler

        # ── Register GGUF extension handler ───────────────────
        def _register_gguf():
            import shared.qtypes.gguf  # noqa: F401  — registers the handler
            print("[INIT] GGUF handler registered with mmgp")
        _register_gguf()

        # ── Patch config loader for GGUF files ────────────────
        def _patch_ltx2_config_loading():
            import models.ltx2.ltx2 as ltx2_mod

            def _patched(path, fallback_config_path=None):
                from mmgp import quant_router
                if isinstance(path, (list, tuple)):
                    path = path[0] if path else ""
                if not path:
                    return {}
                try:
                    _, metadata = quant_router.load_metadata_state_dict(path)
                    if metadata:
                        config_raw = metadata.get("config")
                        if config_raw:
                            config = ltx2_mod._normalize_config(config_raw)
                            if config:
                                return config
                except Exception as e:
                    print(f"[GGUF Patch] metadata read: {type(e).__name__} — using fallback")
                if fallback_config_path and os.path.isfile(fallback_config_path):
                    try:
                        with open(fallback_config_path, "r", encoding="utf-8") as f:
                            config = ltx2_mod._normalize_config(json.load(f))
                            if config:
                                print(f"[GGUF Patch] config from {os.path.basename(fallback_config_path)}")
                                return config
                    except Exception:
                        pass
                return {}

            ltx2_mod._load_config_from_checkpoint = _patched
            print("[INIT] GGUF config patch applied")

        _patch_ltx2_config_loading()

        # ── Locate model files ────────────────────────────────
        transformer_path = os.path.join(MODELS_DIR, TRANSFORMER_DEST)
        if not os.path.isfile(transformer_path):
            raise FileNotFoundError(
                f"Transformer not found: {transformer_path}\n"
                f"Run download_models() first."
            )

        gemma_dir   = os.path.join(MODELS_DIR, GEMMA_FOLDER)
        gemma_files = sorted(glob.glob(os.path.join(gemma_dir, "*.safetensors")))
        quanto_files = [f for f in gemma_files if "quanto" in f]
        text_encoder_file = quanto_files[0] if quanto_files else (gemma_files[0] if gemma_files else None)
        if not text_encoder_file:
            raise FileNotFoundError(f"No Gemma safetensors in {gemma_dir}")

        print(f"[INIT] Transformer  : {os.path.basename(transformer_path)}")
        print(f"[INIT] Text encoder : {os.path.basename(text_encoder_file)}")

        # ── Build model_def for distilled pipeline ────────────
        base_model_type = "ltx2_22B"
        model_def = {"ltx2_pipeline": "distilled"}
        extra = family_handler.query_model_def(base_model_type, model_def)
        model_def.update(extra)

        # ── BF16 for maximum throughput on all sm_80+ GPUs ────
        MODEL_DTYPE = torch.bfloat16
        VAE_DTYPE   = torch.bfloat16

        # ── Load model ────────────────────────────────────────
        print("[INIT] Loading LTX-2.3 22B Distilled 1.1 Q4_K_M ...")
        self.ltx2_model, self.pipe = family_handler.load_model(
            model_filename=transformer_path,
            model_type="ltx2_22B_distilled",
            base_model_type=base_model_type,
            model_def=model_def,
            dtype=MODEL_DTYPE,
            VAE_dtype=VAE_DTYPE,
            text_encoder_filename=text_encoder_file,
        )

        # ── Automatic VRAM-optimised mmgp budgets ─────────────
        # HIGH_PERF  (H200 141GB): maximise VRAM residency, no CPU offload
        # BALANCED   (H100/A100 80GB): generous but allows some offloading
        # EFFICIENT  (RTX PRO 6000 / <60 GB): conservative, more offloading
        if vram_mode == "HIGH_PERF":
            budgets = {
                "transformer":       100000,  # H200: keep entire Q4 transformer in VRAM
                "text_encoder":       20000,
                "video_encoder":      12000,
                "video_decoder":      14000,
                "audio_encoder":       6000,
                "audio_decoder":       6000,
                "vocoder":             5000,
                "spatial_upsampler":  10000,
                "vae":                 8000,
                "*":                   6000,
            }
            vae_tile_size = 0   # no tiling needed on H200
            print(f"[INIT] mmgp budgets: HIGH_PERF (H200 141GB optimised)")
        elif vram_mode == "BALANCED":
            budgets = {
                "transformer":        60000,  # H100/A100: generous VRAM allocation
                "text_encoder":       16000,
                "video_encoder":       8000,
                "video_decoder":      10000,
                "audio_encoder":       4000,
                "audio_decoder":       4000,
                "vocoder":             3000,
                "spatial_upsampler":   6000,
                "vae":                 4000,
                "*":                   4000,
            }
            vae_tile_size = 0   # 80 GB is sufficient for 1080p without tiling
            print(f"[INIT] mmgp budgets: BALANCED (H100/A100 80GB optimised)")
        else:  # EFFICIENT
            budgets = {
                "transformer":        30000,
                "text_encoder":       12000,
                "video_encoder":       6000,
                "video_decoder":       8000,
                "audio_encoder":       3000,
                "audio_decoder":       3000,
                "vocoder":             2000,
                "spatial_upsampler":   4000,
                "vae":                 3000,
                "*":                   2000,
            }
            vae_tile_size = 256   # tile VAE for lower-VRAM cards
            print(f"[INIT] mmgp budgets: EFFICIENT (RTX PRO 6000 / <60 GB)")

        self._vae_tile_size = vae_tile_size

        print("[INIT] Applying mmgp Profile 4 with auto-detected budgets ...")
        offload.profile(
            self.pipe,
            profile_no=4,
            quantizeTransformer=False,
            convertWeightsFloatTo=torch.bfloat16,
            budgets=budgets,
        )
        offload.shared_state["_attention"] = ATTENTION_MODE

        # ── Warm up GC ────────────────────────────────────────
        gc.collect()
        torch.cuda.empty_cache()

        elapsed = time.time() - t0
        vram_used = (
            torch.cuda.get_device_properties(0).total_memory
            - torch.cuda.mem_get_info(0)[0]
        ) / 1024**3
        print(f"\n[INIT] ✅ Model loaded in {elapsed:.1f}s")
        print(f"[INIT] VRAM used after load : {vram_used:.1f} GB / {vram_total:.1f} GB total")
        print(f"[INIT] VAE tile size        : {vae_tile_size} (0 = no tiling)")

        # Store dtype for inference cell
        self._model_dtype = MODEL_DTYPE


print("✅ LTXVideoService Modal class defined (singleton with @modal.enter).")
print(f"   GPU target : {GPU_TYPE}  →  {MODAL_GPU}")
print("   VRAM auto-detection: HIGH_PERF / BALANCED / EFFICIENT")
````

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 5 — WAN2GP / MMGP HELPER UTILITIES           ║
# ╚══════════════════════════════════════════════════════════════╝
#
# Pure-Python helpers that run inside the Modal container.
# Hardware-agnostic: no GPU-specific branches here.

# ─── Resolution helper ────────────────────────────────────────
def _get_resolution(base_res_str: str, aspect_ratio_str: str):
    """Convert UI resolution/aspect strings → (width, height) snapped to 32."""
    base_map = {"1080p": 1088, "720p": 704, "540p": 544, "480p": 480}
    ratio_map = {
        "16:9 Landscape": 16 / 9,
        "4:3 Standard":   4 / 3,
        "1:1 Square":     1.0,
        "3:4 Portrait":   3 / 4,
        "9:16 Portrait":  9 / 16,
    }
    base  = base_map.get(base_res_str, 1088)
    ratio = ratio_map.get(aspect_ratio_str, 16 / 9)
    if ratio >= 1.0:
        height = base
        width  = int(base * ratio)
    else:
        width  = base
        height = int(base / ratio)
    return (width // 32) * 32, (height // 32) * 32


# ─── Frame count helper ───────────────────────────────────────
def _snap_to_ltx_frames(duration_sec: float, fps: float = 24.0, max_frames: int = 721) -> int:
    """Round duration → nearest valid LTX frame count (8k+1 rule).
    LTX distilled valid counts: 1, 9, 17, 25, …  (i.e. 8k+1).
    Floors at 49 frames (2 seconds) and caps at max_frames.
    """
    raw = duration_sec * fps
    k   = max(0, round((raw - 1) / 8))
    return int(max(49, min(8 * k + 1, max_frames)))


# ─── Duration dropdown → frame count ─────────────────────────
DURATION_MAP = {
    "2 Seconds (49 frames)":   49,
    "3 Seconds (73 frames)":   73,
    "5 Seconds (121 frames)":  121,
    "8 Seconds (193 frames)":  193,
    "10 Seconds (241 frames)": 241,
    "15 Seconds (361 frames)": 361,
    "20 Seconds (481 frames)": 481,
    "30 Seconds (721 frames)": 721,
}


# ─── Logging helper ───────────────────────────────────────────
def _log_generation_start(mode, width, height, num_frames, fps, seed, prompt, guide_scale):
    import psutil, torch, time
    free_vram = torch.cuda.mem_get_info(0)[0] / 1024**3
    used_vram = (
        torch.cuda.get_device_properties(0).total_memory - torch.cuda.mem_get_info(0)[0]
    ) / 1024**3
    total_vram = torch.cuda.get_device_properties(0).total_memory / 1024**3
    ram = psutil.virtual_memory()
    print(f"\n{'='*70}")
    print(f"[GEN] Mode       : {mode}")
    print(f"[GEN] Resolution : {width}x{height}  |  {num_frames} frames @ {fps}fps  ({num_frames/fps:.1f}s)")
    print(f"[GEN] Seed       : {seed}")
    print(f"[GEN] Guide      : {guide_scale}")
    print(f"[GEN] GPU        : {torch.cuda.get_device_name(0)}")
    print(f"[GEN] VRAM       : {used_vram:.1f} GB used / {free_vram:.1f} GB free / {total_vram:.1f} GB total")
    print(f"[GEN] RAM        : {ram.used/1024**3:.1f} GB used / {ram.available/1024**3:.1f} GB free")
    print(f"[GEN] Prompt     : {prompt[:140]}{'...' if len(prompt) > 140 else ''}")
    print(f"{'='*70}")


print("✅ Wan2GP / mmgp helper utilities defined.")
````

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 6 — INFERENCE: generate_video()               ║
# ╚══════════════════════════════════════════════════════════════╝
#
# Adds the production-grade generate_video() method to
# LTXVideoService.  Supports T2V / I2V-start / I2V-end /
# I2V first+last frame.  Audio mux included when the model
# generates native audio.
# VAE tile size is automatically set by load_models() based on
# detected VRAM (0 = no tiling for H200/H100/A100 80GB).

import modal

@modal.method()
def generate_video(
    self,
    prompt:         str   = "A cinematic aerial shot of mountains at golden hour, 4K",
    width:          int   = DEFAULT_WIDTH,
    height:         int   = DEFAULT_HEIGHT,
    fps:            float = DEFAULT_FPS,
    num_frames:     int   = DEFAULT_FRAMES,
    seed:           int   = -1,
    guidance_scale: float = DEFAULT_GUIDE,
    num_steps:      int   = DEFAULT_STEPS,
    image_start:    bytes = None,   # raw PNG/JPEG bytes for first frame
    image_end:      bytes = None,   # raw PNG/JPEG bytes for last frame
):
    """Generate a video.  Returns dict with output_path, job_id, stats."""
    import gc
    import io
    import os
    import random
    import subprocess
    import tempfile
    import time
    import numpy as np
    import torch
    import psutil
    from PIL import Image
    from shared.utils.audio_video import save_video

    t_start = time.time()

    # ── Clamp / snap parameters ───────────────────────────────
    width      = max(256, (width  // 32) * 32)
    height     = max(256, (height // 32) * 32)
    num_frames = max(49, min(int(num_frames), 721))
    # Snap to 8k+1 rule
    k = round((num_frames - 1) / 8)
    num_frames = 8 * k + 1

    if seed is None or seed < 0:
        seed = random.randint(0, 2**32 - 1)
    seed = int(seed)

    # ── Detect generation mode ───────────────────────────────
    pil_start = None
    pil_end   = None
    if image_start is not None:
        pil_start = Image.open(io.BytesIO(image_start)).convert("RGB")
    if image_end is not None:
        pil_end = Image.open(io.BytesIO(image_end)).convert("RGB")

    if pil_start is None and pil_end is None:
        mode = "T2V"
    elif pil_start is not None and pil_end is not None:
        mode = "I2V_first_last"
    elif pil_start is not None:
        mode = "I2V_start"
    else:
        mode = "I2V_end"

    _log_generation_start(mode, width, height, num_frames, fps, seed, prompt, guidance_scale)

    # ── VAE tile size: determined at startup by VRAM detection ─
    # self._vae_tile_size is 0 for HIGH_PERF/BALANCED (no tiling)
    # and 256 for EFFICIENT (lower-VRAM) mode.
    vae_tile_size = getattr(self, "_vae_tile_size", 0)

    # ── Step-level progress callback ─────────────────────────
    total_steps   = [num_steps]
    current_step  = [0]
    current_pass  = [1]
    t_step_start  = [time.time()]

    def diffusion_callback(step, latent, is_start,
                           override_num_inference_steps=None,
                           pass_no=None, **kwargs):
        if is_start:
            if override_num_inference_steps is not None:
                total_steps[0] = override_num_inference_steps
            if pass_no is not None:
                current_pass[0] = pass_no
            current_step[0] = 0
            t_step_start[0] = time.time()
            return
        current_step[0] += 1
        stage = (
            "Stage-1 half-res" if current_pass[0] == 1
            else "Stage-2 full-res" if current_pass[0] == 2
            else "Diffusion"
        )
        free_v = torch.cuda.mem_get_info(0)[0] / 1024**3
        dt     = time.time() - t_step_start[0]
        print(
            f"  [{stage}] step {current_step[0]}/{total_steps[0]}"
            f"  +{dt:.1f}s  VRAM free: {free_v:.1f} GB"
        )
        t_step_start[0] = time.time()

    # ── Stage progress callback ──────────────────────────────
    def set_progress_status(status: str):
        print(f"  [STATUS] {status}")

    # ── Build generation kwargs ──────────────────────────────
    gen_kwargs = dict(
        input_prompt=prompt,
        image_start=pil_start,
        height=height,
        width=width,
        frame_num=num_frames,
        fps=fps,
        seed=seed,
        callback=diffusion_callback,
        VAE_tile_size=vae_tile_size,
        input_video_strength=1.0,
        denoising_strength=1.0,
        guide_scale=float(guidance_scale),
        sampling_steps=int(num_steps),
        guide_phases=2,
        n_prompt="",
        enhance_prompt=False,
        video_prompt_type="",
        audio_prompt_type="",
        set_progress_status=set_progress_status,
    )
    if pil_end is not None:
        gen_kwargs["image_end"] = pil_end

    # ── Run generation ────────────────────────────────────────
    gc.collect()
    torch.cuda.empty_cache()
    torch.cuda.synchronize()

    print("[GEN] Diffusion starting ...")
    with torch.inference_mode():
        result = self.ltx2_model.generate(**gen_kwargs)

    t_gen = time.time() - t_start
    print(f"[GEN] ✅ Pipeline complete in {t_gen:.1f}s")

    # ── Unpack result ────────────────────────────────────────
    if isinstance(result, dict):
        video_tensor = result.get("x")
        audio_data   = result.get("audio")
        audio_sr     = result.get("audio_sampling_rate", 24000)
    elif isinstance(result, tuple):
        video_tensor = result[0]
        audio_data   = result[1] if len(result) > 1 else None
        audio_sr     = result[2] if len(result) > 2 else 24000
    else:
        video_tensor = result
        audio_data, audio_sr = None, 24000

    if video_tensor is None or not torch.is_tensor(video_tensor):
        raise RuntimeError(f"No video tensor returned. Got: {type(video_tensor)}")

    print(f"[GEN] Video tensor: {video_tensor.shape}  dtype={video_tensor.dtype}")
    video_tensor = video_tensor.cpu()
    gc.collect()
    torch.cuda.empty_cache()

    # ── Save video ───────────────────────────────────────────
    import uuid
    job_id   = str(uuid.uuid4())[:8]
    out_name = f"ltx_{job_id}_{width}x{height}_{num_frames}f.mp4"
    out_path = os.path.join(OUTPUTS_DIR, out_name)
    os.makedirs(OUTPUTS_DIR, exist_ok=True)

    video_for_save = video_tensor.unsqueeze(0).float() / 127.5 - 1.0
    save_video(
        tensor=video_for_save,
        save_file=out_path,
        fps=fps,
        normalize=True,
        value_range=(-1, 1),
    )
    print(f"[GEN] Video saved: {out_path}")

    # ── Mux native audio (if generated) ─────────────────────
    if audio_data is not None:
        try:
            import soundfile as sf
            audio_tmp = tempfile.mktemp(suffix=".wav")
            if isinstance(audio_data, np.ndarray):
                audio_np = audio_data
                if audio_np.ndim == 2 and audio_np.shape[0] <= 2:
                    audio_np = audio_np.T
                sf.write(audio_tmp, audio_np, int(audio_sr or 24000))
            elif torch.is_tensor(audio_data):
                import torchaudio
                cpu_audio = audio_data.cpu().float()
                if cpu_audio.dim() == 1:
                    cpu_audio = cpu_audio.unsqueeze(0)
                if cpu_audio.dim() == 3:
                    cpu_audio = cpu_audio.squeeze(0)
                torchaudio.save(audio_tmp, cpu_audio, int(audio_sr or 24000))
            else:
                raise ValueError(f"Unknown audio type: {type(audio_data)}")

            final_path = out_path.replace(".mp4", "_audio.mp4")
            subprocess.run(
                [
                    "ffmpeg", "-y",
                    "-i", out_path, "-i", audio_tmp,
                    "-c:v", "copy", "-c:a", "aac", "-b:a", "192k",
                    "-shortest", final_path,
                ],
                check=True,
                capture_output=True,
            )
            if os.path.exists(final_path) and os.path.getsize(final_path) > 0:
                os.remove(out_path)
                out_path = final_path
                out_name = os.path.basename(final_path)
                print("[GEN] ✅ Native audio muxed into video")
            else:
                print("[GEN] ⚠️  Audio mux empty — keeping video-only file")
        except Exception as ae:
            print(f"[GEN] ⚠️  Audio mux failed: {ae}")

    # ── Cleanup old outputs ──────────────────────────────────
    try:
        import glob as _glob
        old = sorted(
            _glob.glob(os.path.join(OUTPUTS_DIR, "*.mp4")),
            key=os.path.getmtime,
        )
        for old_f in old[:-MAX_OUTPUTS_KEEP]:
            os.remove(old_f)
            print(f"[GEN] Cleaned old output: {os.path.basename(old_f)}")
    except Exception:
        pass

    output_volume.commit()

    # ── Final VRAM / RAM stats ───────────────────────────────
    vram_used = (
        torch.cuda.get_device_properties(0).total_memory
        - torch.cuda.mem_get_info(0)[0]
    ) / 1024**3
    ram_used  = psutil.virtual_memory().used / 1024**3
    t_total   = time.time() - t_start

    stats = {
        "generation_time_s": round(t_total, 1),
        "vram_used_gb":       round(vram_used, 2),
        "ram_used_gb":        round(ram_used, 2),
        "resolution":         f"{width}x{height}",
        "frame_count":        num_frames,
        "fps":                fps,
        "duration_s":         round(num_frames / fps, 2),
        "mode":               mode,
        "seed":               seed,
    }
    print(f"[GEN] Stats: {stats}")
    del video_tensor, video_for_save
    gc.collect()
    torch.cuda.empty_cache()

    return {
        "job_id":      job_id,
        "output_path": out_path,
        "output_name": out_name,
        "stats":       stats,
    }

# Inject method into class
LTXVideoService.generate_video = generate_video

print("✅ generate_video() method defined and injected into LTXVideoService.")
````

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 7 — FASTAPI SCHEMAS & MODELS                  ║
# ╚══════════════════════════════════════════════════════════════╝

from pydantic import BaseModel, Field
from typing import Optional


class VideoGenerateRequest(BaseModel):
    """Request body for POST /generate"""
    prompt:         str   = Field(
        default="A cinematic aerial shot of mountains at golden hour, 4K",
        description="Text prompt describing the video to generate.",
    )
    width:          int   = Field(default=DEFAULT_WIDTH,  ge=256, le=3840)
    height:         int   = Field(default=DEFAULT_HEIGHT, ge=256, le=2160)
    fps:            float = Field(default=DEFAULT_FPS,   ge=8.0, le=60.0)
    num_frames:     int   = Field(default=DEFAULT_FRAMES, ge=49,  le=721)
    seed:           int   = Field(default=-1, description="-1 = random")
    guidance_scale: float = Field(default=DEFAULT_GUIDE, ge=1.0, le=10.0)
    num_steps:      int   = Field(default=DEFAULT_STEPS, ge=1,   le=50)
    # Optional image bytes encoded as base64 strings
    image_start_b64: Optional[str] = Field(
        default=None,
        description="Base-64 encoded PNG/JPEG bytes for the first frame (Image-to-Video).",
    )
    image_end_b64: Optional[str] = Field(
        default=None,
        description="Base-64 encoded PNG/JPEG bytes for the last frame (First+Last mode).",
    )


class GenerationStats(BaseModel):
    generation_time_s: float
    vram_used_gb:      float
    ram_used_gb:       float
    resolution:        str
    frame_count:       int
    fps:               float
    duration_s:        float
    mode:              str
    seed:              int


class VideoGenerateResponse(BaseModel):
    """Response body from POST /generate"""
    job_id:      str
    video_url:   str   = Field(description="Relative URL to download the generated video.")
    output_name: str
    stats:       GenerationStats


class HealthResponse(BaseModel):
    status:       str
    gpu:          str
    vram_free_gb: float
    vram_total_gb: float
    vram_mode:    str
    cuda_version: str
    torch_version: str


print("✅ FastAPI Pydantic schemas defined:")
print("   • VideoGenerateRequest")
print("   • VideoGenerateResponse")
print("   • GenerationStats")
print("   • HealthResponse  (extended with vram_total_gb, vram_mode, cuda_version, torch_version)")
````

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 8 — HTTP ENDPOINT  POST /generate             ║
# ╚══════════════════════════════════════════════════════════════╝

import modal
from fastapi import FastAPI, HTTPException, Response
from fastapi.responses import FileResponse
import base64

web_app = FastAPI(
    title="LTX-2.3 22B Video Generation API",
    description=(
        "Production endpoint for LTX-2.3 22B Distilled 1.1 Q4_K_M\n"
        "running on Modal with Wan2GP + mmgp.\n"
        "Hardware-adaptive: H200 141GB · H100 80GB · A100 80GB · RTX PRO 6000."
    ),
    version="2.0.0",
)


@web_app.get("/health", response_model=HealthResponse)
async def health_check():
    """Liveness / GPU health probe with full hardware report."""
    import torch
    if not torch.cuda.is_available():
        raise HTTPException(status_code=503, detail="No CUDA device")
    free_vram  = torch.cuda.mem_get_info(0)[0] / 1024**3
    total_vram = torch.cuda.get_device_properties(0).total_memory / 1024**3
    if total_vram > VRAM_HIGH_PERF_THRESHOLD_GB:
        vram_mode = "HIGH_PERF"
    elif total_vram > VRAM_BALANCED_THRESHOLD_GB:
        vram_mode = "BALANCED"
    else:
        vram_mode = "EFFICIENT"
    return HealthResponse(
        status="ok",
        gpu=torch.cuda.get_device_name(0),
        vram_free_gb=round(free_vram, 2),
        vram_total_gb=round(total_vram, 2),
        vram_mode=vram_mode,
        cuda_version=torch.version.cuda or "unknown",
        torch_version=torch.__version__,
    )


@web_app.post("/generate", response_model=VideoGenerateResponse)
async def generate_endpoint(req: VideoGenerateRequest):
    """Generate a video.

    - **prompt**: text description of the video
    - **image_start_b64**: optional base-64 first frame (enables Image-to-Video)
    - **image_end_b64**: optional base-64 last frame (enables First+Last mode)
    - Returns job_id and video_url to download the result
    """
    # Decode optional image bytes
    image_start_bytes = None
    image_end_bytes   = None
    if req.image_start_b64:
        try:
            image_start_bytes = base64.b64decode(req.image_start_b64)
        except Exception:
            raise HTTPException(status_code=422, detail="image_start_b64 is not valid base64")
    if req.image_end_b64:
        try:
            image_end_bytes = base64.b64decode(req.image_end_b64)
        except Exception:
            raise HTTPException(status_code=422, detail="image_end_b64 is not valid base64")

    # Call the Modal class method
    try:
        svc    = LTXVideoService()
        result = svc.generate_video.remote(
            prompt=req.prompt,
            width=req.width,
            height=req.height,
            fps=req.fps,
            num_frames=req.num_frames,
            seed=req.seed,
            guidance_scale=req.guidance_scale,
            num_steps=req.num_steps,
            image_start=image_start_bytes,
            image_end=image_end_bytes,
        )
    except Exception as e:
        raise HTTPException(status_code=500, detail=str(e))

    return VideoGenerateResponse(
        job_id=result["job_id"],
        video_url=f"/download/{result['output_name']}",
        output_name=result["output_name"],
        stats=GenerationStats(**result["stats"]),
    )


@web_app.get("/download/{filename}")
async def download_video(filename: str):
    """Download a generated video by filename."""
    import os
    path = os.path.join(OUTPUTS_DIR, filename)
    if not os.path.isfile(path):
        raise HTTPException(status_code=404, detail=f"File not found: {filename}")
    return FileResponse(path, media_type="video/mp4", filename=filename)


@web_app.get("/outputs")
async def list_outputs():
    """List all generated output files."""
    import os, glob
    files = sorted(glob.glob(os.path.join(OUTPUTS_DIR, "*.mp4")), key=os.path.getmtime, reverse=True)
    return {
        "count": len(files),
        "files": [
            {
                "name":    os.path.basename(f),
                "size_mb": round(os.path.getsize(f) / 1024**2, 1),
                "url":     f"/download/{os.path.basename(f)}",
            }
            for f in files
        ],
    }


# ── Attach FastAPI app to Modal ───────────────────────────────
@app.function(
    gpu=MODAL_GPU,
    cpu=8,
    memory=65536,
    timeout=3600,
    volumes={
        MODELS_DIR:  model_volume,
        OUTPUTS_DIR: output_volume,
    },
    concurrency_limit=1,
)
@modal.asgi_app()
def fastapi_app():
    return web_app


print("✅ FastAPI endpoints registered:")
print("   GET  /health          (full hardware report)")
print("   POST /generate")
print("   GET  /download/{filename}")
print("   GET  /outputs")
````

````python
# ╔══════════════════════════════════════════════════════════════╗
# ║         CELL 9 — OUTPUT MANAGEMENT & LOGGING               ║
# ╚══════════════════════════════════════════════════════════════╝

import modal

@app.function(
    volumes={
        MODELS_DIR:  model_volume,
        OUTPUTS_DIR: output_volume,
    },
    timeout=300,
    cpu=2,
)
def manage_outputs(keep_last: int = MAX_OUTPUTS_KEEP):
    """List outputs, log stats, and clean up old files keeping only the last N."""
    import os
    import glob
    import time

    files = sorted(
        glob.glob(os.path.join(OUTPUTS_DIR, "*.mp4")),
        key=os.path.getmtime,
    )

    total_size = sum(os.path.getsize(f) for f in files) / 1024**3
    print(f"[OUTPUTS] {len(files)} files  |  Total size: {total_size:.2f} GB")

    for f in files:
        mtime = time.strftime("%Y-%m-%d %H:%M", time.localtime(os.path.getmtime(f)))
        size  = os.path.getsize(f) / 1024**2
        print(f"  {mtime}  {size:7.1f} MB  {os.path.basename(f)}")

    to_delete = files[:-keep_last] if len(files) > keep_last else []
    for f in to_delete:
        os.remove(f)
        print(f"  [DELETED] {os.path.basename(f)}")

    if to_delete:
        output_volume.commit()
        print(f"[OUTPUTS] Cleaned {len(to_delete)} old files. Volume committed.")
    else:
        print("[OUTPUTS] No cleanup needed.")

    return {
        "total_files":    len(files),
        "total_size_gb":  round(total_size, 3),
        "deleted":        len(to_delete),
    }


print("✅ manage_outputs() Modal function defined.")
print("   Run with:  modal run ltx_modal_h200.ipynb::manage_outputs")
````

### Automatic VRAM Mode Selection

At container startup, `load_models()` detects available VRAM and selects the optimal memory profile **automatically** — no configuration needed:

| Detected VRAM | Mode | GPU | Behaviour |
|---------------|------|-----|-----------|
| > 120 GB | **HIGH_PERF** | H200 141GB | Maximum VRAM budgets, no CPU offloading, no VAE tiling |
| > 60 GB | **BALANCED** | H100 / A100 80GB | Generous budgets, minimal offloading, no VAE tiling |
| ≤ 60 GB | **EFFICIENT** | RTX PRO 6000 etc. | Conservative budgets, VAE tiling enabled |
